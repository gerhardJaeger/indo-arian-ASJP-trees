```r
library(ggplot2)
library(bayesplot)
```

```r
posterior = read.table('output/gamma11_0.p', header=T, check.names=F)

head(posterior)
```

```r
df = data.frame(x=posterior[,'rates[11]'] - posterior[,'rates[7]']) # r(-clf -> +clf|opt) - r(-clf -> +clf|obl)
```

```r
ggplot(df, aes(x=x, y=..density..)) + 
    geom_histogram(alpha=.3) + 
    geom_density(col='red', fill='red', alpha=.5)
```
